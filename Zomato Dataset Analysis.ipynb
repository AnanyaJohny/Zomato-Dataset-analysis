```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df=pd.read_csv('/Users/ananyajohny/Downloads/zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Cost2plates'].unique()
```

```python
def handleCost2plates(value):
    if ',' in value:
        x=str(value).split(',')
        value=float(x[0]+x[1])
        return value
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(handleCost2plates)
df.head()
        
```

```python
df.info()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handleRestTypes(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handleRestTypes)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location_lessthan300=df['location'].value_counts()
```

```python
location_lessthan300=location_lessthan300[location_lessthan300<300]
```

```python
location_lessthan300
```

```python
def handleLocation(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handleLocation)
```

```python
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines_lessthan100=df['cuisines'].value_counts()
```

```python
cuisines_lessthan100=cuisines_lessthan100[cuisines_lessthan100<100]
```

```python
cuisines_lessthan100
```

```python
def handleCuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handleCuisines)
```

```python
df['cuisines'].value_counts()
```
